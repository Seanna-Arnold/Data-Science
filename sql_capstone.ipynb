# CREATING DATABASE

```python
#databasetype+connectionmethod://user:password@external_url:port/database_name
engine = create_engine('postgresql+psycopg2://seanna:1@127.0.0.1:5432/mydb')
```

```python
#create database
datatable = {
    'id':[3001,3002,3003,3004,3005]
    ,'Title':['A Tale of Two Cities', "Harry Potter and the Philosopher's Stone", 'The Lion, the Witch and the Wardrobe', 'The Lord of the Rings', 'Alice in Wonderland']
    ,'Auther':['Charled Dickens', 'J.K. Rowling', 'C.S. Lewis', 'J.R.R. Tolkien', 'Lewis Carroll']
    ,'Qty':[30, 40, 25, 37, 12]
}
```

```python
df = pd.DataFrame.from_dict(datatable)
df.to_sql('data', engine.connect(), if_exists='replace')
```

```python
meta = MetaData()

data = Table(
   'data', meta, 
   Column('id', Integer, primary_key = True), 
   Column('Title', String), 
   Column('Auther', String),
   Column('Qty', Integer))
```

```python
class Bookstore:
    """
    This class would be used to manage a book stores inventory system.
    
    arguments:
        databasetype (str): This is the type of database you're connecting to
        connectionmethod (str):
        user (str):
        password (str):
        external_url (str):
        port (str):
        database_name (str):
        
    returns:
        n/a
    
    """
    def __init__(self, databasetype, connectionmethod, user, password, external_url, port, database_name):
        self.engine = create_engine(f'{databasetype}+{connectionmethod}://{user}:{password}@{external_url}:{port}/{database_name}')

    def enter_book(self):
        UID = int(input('Please input id: '))
        table_a = input('Please input Title: ').capitalize()
        name_b = input('Please input the author: ').capitalize()
        quantity = int(input('Please input the quantity: '))
        
        to_insert = {
            'id':[UID]
            ,'Title':[table_a]
            ,'Auther':[name_b]
            ,'Qty':[quantity]}
        pd.DataFrame.from_dict(to_insert).to_sql('data', self.engine.connect(), if_exists='append')
        
        return 'books inserted'
    
    def update_book(self):
        """
        This class would be used to manage a book stores inventory system.

        arguments:
            N/A

        returns:
            (str): "Not a valid column." if successfully ran.
        """
        #User must select column and row to search by
        #If the column the user inputs is not a valid row tell them
        update = input("What column to query: ")
        if update == "id":
            value = int(input('Please input id: '))
        elif update == "Title":
            value = input('Please input Title: ').capitalize()
        elif update == "Auther":
            value = input('Please input the author: ').capitalize()
        elif update == "Qty":
            value = int(input('Please input the quantity: '))
        else:
            return "Not a valid column."
        
        
        #let user select what column to update and with what value
        """
        This class would be used to manage a book stores inventory system.

        arguments:
            N/A

        returns:
            (str):"Your column was successfully updated." if successfully ran.
        """
        column_to_update = input("What column to update by: ")
        if column_to_update == "id":
            value_to_update = int(input('Please input updated id: '))
        elif column_to_update == "Title":
            value_to_update = input('Please input updated Title: ').capitalize()
        elif column_to_update == "Auther":
            value_to_update = input('Please input the updated author: ').capitalize()
        elif column_to_update == "Qty":
            value_to_update = int(input('Please input the updated quantity: '))
        else:
            return "Not a valid column."
        
        Session = sessionmaker(bind = engine)
        session = Session()

        session.query(data).filter(data.columns[f'{update}']==value).update({f"{column_to_update}": value_to_update})
        session.commit()
        return "Your column was successfully updated."
    
    
    def delete_book(self):
        """
        This class would be used to manage a book stores inventory system.

        arguments:
            N/A

        returns:
            (str): 'Row has been deleted.' if successfully ran.
        """
        delete_column = input("What column to query for deletion: ")
        if delete_column == "id":
            value = int(input('Please input id: '))
        elif delete_column == "Title":
            value = input('Please input Title: ').capitalize()
        elif delete_column == "Auther":
            value = input('Please input the author: ').capitalize()
        elif delete_column == "Qty":
            value = int(input('Please input the quantity: '))
        else:
            return "Not a valid column."
        
        Session = sessionmaker(bind = engine)
        session = Session()

        session.query(data).filter(data.columns[f'{delete_column}']==value).delete()
        session.commit()
        return 'Row has been deleted.'
    
    
    def search_books(self):
        """
        This class would be used to manage a book stores inventory system.

        arguments:
            N/A

        returns:
            The searched item.
        """
        search = input("What column to search by: ")
        if search == "id":
            value = int(input('Please input id: '))
        elif search == "Title":
            value = input('Please input Title: ').capitalize()
        elif search == "Auther":
            value = input('Please input the author: ').capitalize()
        elif search == "Qty":
            value = int(input('Please input the quantity: '))
        else:
            return "Not a valid column."
        
        Session = sessionmaker(bind = self.engine)
        session = Session()
        root = session.query(data).filter(data.columns[f'{search}']==value)

        return root.all()
    
    def exit(self):
        return 'user exited'
    
    def run_app(self):
        """
        This class would be used to manage a book stores inventory system.

        arguments:
            N/A

        returns:
            Menu
        """
        user_selection = int(input('what would you like to do: '))
        
        if user_selection == 1:
            output = self.enter_book()
        elif user_selection == 2:
            output = self.update_book()
        elif user_selection == 3:
            output = self.delete_book()
        elif user_selection == 4:
            output = self.search_books()
        elif user_selection == 0:
            output = self.exit()
        else:
            return f'sorry {user_selection} is not an option'
        
        
        return output
```
